```python
import json
```

```python
with open('credits_list.json') as f:
    credits_list = json.load(f)
print(credits_list)
```

```python
title_dict = {credit['credit_id']: credit['title'] for credit in credits_list}
print(title_dict)
```

```python
director_titles_dict = {credit['credit_id']: credit['title'] for credit in credits_list if credit['job'] == 'Director'}
print(director_titles_dict)
```
